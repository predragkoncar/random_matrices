```python
from skrmt.ensemble import CircularEnsemble
import matplotlib.pyplot as plt
```

# Circular Unitary Ensemble (CUE)

**FACTS**

---


*   Real and complex part are gaussian distributed.
*   They are invariant under unitary conjugation.
*   Also known as 2-Dyson random matrices (beta=2).

```python
cue = CircularEnsemble(beta=2, n=500)
```

```python
cue.plot_eigval_hist(bins=100)
```

#Circular Orthogonal Ensemble (COE)

**FACTS**

---


*   Real entries are gaussian distributed.
*   They are invariant under orthogonal conjugation.
*   Also known as 1-Dyson random matrices (beta=1).

```python
coe = CircularEnsemble(beta=1, n=4)
```

```python
def matrixPrint(mat, fmt="g"):
    maxC = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]
    for x in mat:
        for i, y in enumerate(x):
            print(("{:"+str(maxC[i])+fmt+"}").format(y), end="  ")
        print("")
```

```python
matrixPrint(coe.matrix)
```

#Circular Symplectic Ensemble (CSE)

**FACTS**

---



*   They are invariant under conjugation by symplectic group.
*   Also known as 4-Dyson random matrices (beta=4).

```python
cse = CircularEnsemble(beta=4, n=500)
cse.plot_eigval_hist(bins=50)
```
